## Algoritmo de búsqueda por anchura en grafos

#### ¿Qué es un grafo?

Un grafo, es una entidad matemática que consiste de un conjunto de nodos, unido mediante vértices. En ciencias de la computación, son una estructura de datos bastante eficiente para el modelado de objetos, problemas que requieren relacionar cosas, descripción de caminos, creación y estudio de redes, etc.

#### BFS

El algoritmo de busqueda por anchura, tiene como función, atravesar todo el grafo, distinguiendo en el proceso, nodos visitados y no visitados, con el fin de encontrar algún nodo, determinar las componentes conexas, el camino más corto, etc.

Este algoritmo es útil para resolver múltiples problemas relacionados a la teoría de grafos, problemas de redes, y problemas relacionados a navegación, aquí algunos ejemplos:

- Encontrar el camino más corto entre dos nodos de un grafo sin peso (Valores numéricos en sus aristas).
- En redes "Peer-to-Peer", se utiliza el BFS para determinar todos los nodos vecinos.
- En redes sociales, para determinar los amigos en común, se utiliza BFS para determinar a cuántas conexiones estamos de una persona.
- En sistemas GPS, se utiliza el BFS para determinar las localizaciones cercanas.
- Detectar un ciclo en un grafo dirigido o no dirigido.
- Encontrar todos los nodos de una componente en un grafo.

Existen muchas más pero estas son de las más importantes. 

```python
from collections import defaultdict 
 
class Graph: # Tipo de dato abstracto para representar el grafo
    # Constructor
    def __init__(self):
        
        # Se crea un diccionario de listas para representar el grafo como una lista de adyacencia
        self.graph = defaultdict(list)
        
    # Agregamos datos al grafo
    def addEdge(self, u, v):
        self.graph[u].append(v)
 
    # Algoritmo a analizar (BFS)
    def BFS(self, s):
 
        # Creamos nuestra lista de visitados
        visited = [False] * (max(self.graph) + 1) # 3m  ||  S: n
 
        # Creamos una cola para visitar los nodos
        queue = [] # 2 || n
 
        # Agregamos el elemento por visitar a la cola y lo marcamos como visitado
        queue.append(s) #   3 || 1
        visited[s] = True # 3 || 1
 
        while queue: #Mientras haya elementos en la cola repetimos 
            # n+1 
 
            # Desencolamos el elemento visitado y lo imprimimos
            s = queue.pop(0) # 3n || 1
            print(s, end=" ") # ---Por definir----- n
 
            # Iteramos por todos los nodos cercanos a nuestro nodo actual y lo agregamos a la cola
            #  si este no es visitado aun
            for i in self.graph[s]: # n+1
                if visited[i] == False:# 3m
                    queue.append(i) # 3 || 1
                    visited[i] = True # 3 || 1
    # 8n + 4m + 8 = O(n+m) n siendo nodos y m siendo aristas, también se puede representar como O(V+E)
    # 2n + 5 = O(n) = O(V)
 
 
# Funcion principal
if __name__ == '__main__':

    entry = int(input('Ingresa la cantidad de entradas para represntar el grafo: '))
    
    # Creamos un grafo
    g = Graph()
    
    for i in range(entry):
        u = int(input('Ingresa el nodo: '))
        v = int(input('Ingresa un nodo adyacente al nodo anterior: '))
        g.addEdge(u, v)
        #g.addEdge(0, 1)
        #g.addEdge(0, 2)
        #g.addEdge(1, 2)
        #g.addEdge(2, 0)
        #g.addEdge(2, 3)
        #g.addEdge(3, 3)
    
    nodoInicio = int(input('Ingresa el nodo en donde quieres iniciar'))
    print('Empezando con el nodo: ' + str(nodoInicio))
    g.BFS(nodoInicio)
 
```

# Algoritmo para evaluar un polinomio de n términos, construido mediante la estrategia fuerza bruta.

Este algoritmo, recibe como parámetros la lista de coeficientes del polinomio y el valor en el que se evaluará.

La lista se llena de izquierda a derecha, siendo coef[0] el término independiente y coef[len(coef)-1] el término con el grado más alto. Si el polinomio no cuenta con algún factor de grado x^n, se indicará que el coeficiente es 0, anulando el término.

El procedimiento consiste en iterar por toda la lista de coeficientes y exponenciar el término a evaluarsegún el término del polinomio que le corresponda, el cuál es determinado por el iterador "i". Una vez terminado el segundo ciclo, se multiplica nuestra x^n por el coeficiente correspondiente, para continuar con el término que sigue aumentando "i".

Una vez terminados los dos ciclos, se devuelve el total.

```python
def evalua_polinomio1(coef, x):
    total = 0                   # 3 || 1
    i = 0                       # 3 || 1
    while i < len(coef):        # 4n+4
        xn = 1.0               # 3n || 1
        j = 0                   # 3n || 1
        while j < i:           # 3(n*n)+3 
            xn *= x             # 4(n*n) || 1
            j += 1              # 4(n*n) || 1
        total += coef[i] * xn   # 6n || 1
        i += 1                  # 4n || 1
    return total                # 1 retorno

# 10n^2 + 20n + 14 = O(n^2)
# La complejidad espacial es O(1)
pol = evalua_polinomio1([1,2,3,4,5],2)
print('El polinomio evaluado resulta en: '+str(pol))
```

# Función Hash para mapeo de contraseñas

El algoritmo hash es una función matemática utilizada para convertir los datos de entrada de cualquier tamaño en un hash o una salida larga fija conocida como hash o resumen. Uno de los usos del hashing es el almacenamiento seguro de contraseñas en bases de datos. En lugar de almacenar la contraseña, esta se hashea y se almacena el resultado.

En nuestro caso particular, será para hashear una cadena de caracteres, por lo que nuestra función, declara una suma, la cual irá desde i = 0, hasta i = Tamaño de la cadena (i siendo nuestro iterador). Por cada valor alfanumérico del caracter se irá sumando su valor, para finalmente aplicarle la operación MOD a este resultado con el fin de obtener el índice de la tabla donde se guardará.



```python
# -*- coding: utf-8 -*-

def hash(cadena):
    sum = 0 # 3 || 1
    for i in cadena: # 3n+1 Se repetirá n+1 veces, n siendo el tamaño de la cadena
        sum += ord(i) # 4n || 1
    clave = sum % 1000 # 4 || 1
    #Vamos a suponer que 1000 es el tamaño de la tabla
    return clave # 1 || 1

# Polinomio: 7n + 9 = O(n)
# Complejidad espacial: O(1)

passw = input("Ingrese la clave a mapear: ")
id = hash(passw)
print('El id de la clave ', passw ,' sería '+ str(id))
```
