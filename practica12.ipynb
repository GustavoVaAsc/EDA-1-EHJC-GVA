### Backtracking.

#### ¿Qué es backtracking?

El backtracking es una técnica de programación que busca entre todo el conjunto de respuestas, tomando decisiones en distintos puntos del problema. En cuanto encontramos que una decisión explorada no es óptima, o viola algún parámetro del problema, efectuamos el "backtrack" a un estado anterior para probar una nueva elección, así hasta encontrar una solución válida.

Usualmente se usa backtracking para resolver problemas de combinaciones y permutaciones, algunos de ellos son:

- El problema de las N reinas
- Llenar un sudoku
- Determinar si los elementos de un subconjunto pueden ser sumados para construir un número N.

#### Algoritmo de búsqueda por profundidad en grafos (Depth First Search DFS).

Como el nombre lo indica, este algoritmo buscará llegar a la profundidad máxima de un grafo dado un nodo como punto de partida, esto mediante atravesar el grafo de manera recursiva. 

Este algoritmo puede ser utilizado para encontrar una solución a un laberinto, para encontrar un ciclo dentro de un grafo, determinar si un grafo es bipartito, encontrar soluciones para problemas de horarios, efectuar un ordenamiento topológico.

DFS, utiliza la técnica de Backtracking, debido a que cuando llega al último nodo de profundidad dentro de una "rama" del grafo, se antecede a la siguiente para determinar si existe otro camino, así hasta que todos los nodos posibles dentro del alcance del nodo a explorar queden visitados.

```python
from collections import defaultdict

graph = {
    5: [3, 7],
    3: [2, 4],
    7: [8],
    2: [],
    4: [8],
    8: []
}

visited = set()

def dfs(visited, graph, node): 
    if node not in visited: # 4 (Esto debido a que el set está implementado sobre una tabla de Hash, su acceso es O(1))
        print (node) # 2
        visited.add(node) # 3
        for neighbour in graph[node]: # Se itera sobre toda la lista de adyacencia del nodo, lo tomaremos como n
                                      # Entonces 4(n+1)
            dfs(visited, graph, neighbour) # Por cada iteracion se hace una llamada recursiva, tomando como 
                                           # Siendo que la cantidad de vertices para el peor caso,
                                           # Tenemos 4(n(n-1)/2)
# Polinomio 2n^2 + 6n + 9 = O(n^2)
print("Ejecutando DFS...")
dfs(visited, graph, 5)
```

### Bottom up.

#### ¿Qué es bottom up?

La estrategia de construcción de algoritmos "Bottom-Up" (de abajo hacia arriba) es un enfoque en el diseño de algoritmos donde se comienza resolviendo subproblemas más pequeños y luego se van combinando para resolver el problema principal de manera gradual. En otras palabras, se empieza desde los elementos más básicos y se construye hacia arriba hasta llegar a la solución completa.

La estrategia "Bottom-Up" es especialmente útil cuando el problema se puede descomponer de manera natural en subproblemas más pequeños y cuando la solución óptima del problema principal puede construirse a partir de las soluciones óptimas de los subproblemas. A menudo, esta estrategia se utiliza en programación dinámica, donde los resultados intermedios se almacenan en una tabla y se reutilizan para calcular soluciones de problemas más grandes.

#### Algoritmo para calcular n numero de la serie de Fibonacci construido de manera iterativa.

En esta versión de Fibonacci se plantea de manera iterativa, de tal modo que ahora nuestros casos bases serán inicializados desde abajo con un arreglo, posteriormente se incrementa este arreglo mediante el bucle for, con las soluciones anteriores, en las posiciones i-1 y i-2.

Como podemos observar, tenemos los casos base inicializados en los dos primeros valores del arreglo, estas nuestras soluciones más pequeñas, nuestro enfoque bottom up se aplica al ir construyendo de abajo hacia arriba los números de forma incremental.


```python
def fibonacci_bottom_up(n):
    fib_nums = [0, 1]  # Casos base (Inicializados en el arreglo) # 5
    for i in range(2, n+1): # Construccion iterativa de la sucesion # 3 (n+2-2) = 3n
        fib_nums.append(fib_nums[i-1] + fib_nums[i-2]) # Construyendo el caso actual con los dos anteriores
                                                       # 5n
    return fib_nums # 1

# Polinomio: 8n + 6 = O(n)


n = 10
fibonacci_sequence = fibonacci_bottom_up(n)
print(fibonacci_sequence)
```
